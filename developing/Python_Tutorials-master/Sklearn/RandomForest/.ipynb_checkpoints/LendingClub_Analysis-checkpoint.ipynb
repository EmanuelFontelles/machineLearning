```python
#Import necessary functions for analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline

#Import random function to downsample for initial analysis
import random

# Loans file to load data from
f = './data/loan.csv'

# Count lines in the loan data
num_lines = sum(1 for l in open(f))

# Create sample size - in this case ~10% (89k rows)
size = int(num_lines / 10)

# Create row indices to skip using random sample
skip_idx = random.sample(range(1, num_lines), num_lines - size)

# Read the loan data sample into a dataframe
loans = pd.read_csv(f, skiprows=skip_idx, low_memory=False)

#loans.loc[:, 'tot_cur_bal':].head()
loans.head()
```

```python
loans.columns
```

```python
#Ask about this - good o initially filter down columns?
#Create list of initial features we might be interested in - based on Data Dictionary information
loans_cols = ['loan_amnt', 'term', 'int_rate', 'installment', 
              'grade','emp_length', 'home_ownership', 'annual_inc',
              'verification_status', 'issue_d', 'loan_status', 'desc',
              'purpose', 'title', 'dti', 'delinq_2yrs', 'inq_last_6mths',
              'open_acc','revol_util', 'out_prncp', 'collections_12_mths_ex_med',
              'mths_since_last_major_derog', 'application_type', 'annual_inc_joint',
              'verification_status_joint', 'acc_now_delinq', 'tot_coll_amt', 'tot_cur_bal']
loans[loans_cols].head()
```

```python
#Check number of rows and columns in the sample
loans.shape
```

```python
#Check column datatypes
loans.dtypes
```

```python
#Check to see which features are missing data from the dataset
loans.info()
```

```python
#Drop columns that contain at least 20% missing values from the sample
loans = loans.loc[:, pd.notnull(loans).sum()>len(loans)*.8]
#Result is 44 columns
loans.info()
```

```python
#Let's remove some other columns we think will not be useful for the analysis (used data dictionary)
#These are based on the loan amount
loans.drop(['funded_amnt', 'funded_amnt_inv']
           , axis=1
           , inplace=True
           , errors = 'ignore')
#These features depends on loan status (collections due to loan status issues)
loans.drop('recoveries'
           , axis=1
           , inplace=True
           , errors = 'ignore')
#Other columns we don't think will be helpful in predicting loan status based on dictionary
loans.drop(['member_id', 'url', 'zip_code', 'addr_state', 'policy_code', 'id']
           , axis=1
           , inplace=True
           , errors = 'ignore')

loans.info()
```

```python
#Check value counts in the loan_status column - this is what we are trying to predict
#Current / fully paid = good, other = bad?
loans.loan_status.value_counts(dropna=False)
```

```python
#Let's make the status column binary (1="Good Standing"; 0="Not good standing") based on status
loans['loan_status'] = loans['loan_status'].apply(lambda x: 1 if x =='Current' else 1 if x =='Fully Paid' else 0)
loans.loan_status.value_counts(dropna=False)

#loans.loan_status = loans['loan_status'].map({'Current': 1, 'Fully Paid': 1,'Charged Off': 0, 'Late (30-120 days)': 1'Current': 1, 'Fully Paid': 1 })
```

```python
#Convert loan term to numeric ("term_months")
#loans.term.value_counts()
loans['term_months'] = loans.term.str.extract('(\d+)',expand=True)
#loans.term_months.value_counts()
#Drop the original term column
loans.drop(['term'], axis=1, inplace=True);
```

```python
#Check values in grade column - we can convert this to numeric
loans.grade.value_counts()
```

```python
#Convert loan grade to numeric ("A=1, B=2, etc")
loans['grade_num'] = [ ord(x) - 64 for x in loans.grade ]
#loans.head()
```

```python
#remove sub grade column
loans.drop(['sub_grade'], axis=1, inplace=True);
```

```python
#Check values in employment length column - see if we can convert to numeric
loans.emp_length.value_counts()
```

```python
#Add employment length numerical - <1 = 1, 10+ = 10
loans['emp_length_yrs'] = loans.emp_length.str.extract('(\d+)',expand=True).astype(float)
#Drop original emp_length column
loans.drop(['emp_length'], axis=1, inplace=True);
loans['emp_length_yrs'].value_counts()
```

```python
#Create dataframe of only numerical columns from the loan data
loans_num = loans.select_dtypes(exclude=['object'])
loans_num.info()
```

```python
#loans_num.info()
#Fill missing values with average value for now - to use sklearn feature selection
loans_num = loans_num.fillna(loans_num.mean())
loans_num.info()
```

```python
#Separate into 2 dataframes X and Y
loans_X = loans_num.iloc[:, loans_num.columns != 'loan_status']
loans_y = loans_num.iloc[:, loans_num.columns == 'loan_status']

#loans_y.head()
loans_X.head()
```

```python
#Let's check the correlation of all the other columns with loan status
loans_X.corrwith(loans_y['loan_status'])
```

```python
#Pick top 5 features by correlation
FeatureCols = ['int_rate', 'inq_last_6mths', 'revol_util', 'annual_inc']
loans_X = loans[FeatureCols]
loans_X.head()

#Join x and y
loans_all = pd.concat([loans_X, loans_y], axis=1)
loans_all.head()
```

```python
#52 numeric columns - notice that credit inquiry info (2184 rows) columns may be co dependent
loans_num.describe()
#Remove columns with correlation close to 0
```

```python
##Attempting to see which features are best
#Use sklearn ExtraTreesClassifier to rank feature importance - help decide which features to keep
from sklearn import datasets
from sklearn import metrics
from sklearn.ensemble import ExtraTreesClassifier

# fit an Extra Trees model to the loan data
model = ExtraTreesClassifier()
model.fit(loans_X, loans_y)
# display the relative importance of each attribute
print(model.feature_importances_)
```

##KNN Classification##

```python
# Create feature matrix (X).
feature_cols = ['int_rate', 'inq_last_6mths', 'revol_util', 'annual_inc']
X = loans[feature_cols]
```

```python
# Create response vector (y).
y = loans['loan_status']
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=99)
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

```python
y_pred_class = knn.predict(X_test)
print(metrics.accuracy_score(y_test, y_pred_class))
```
