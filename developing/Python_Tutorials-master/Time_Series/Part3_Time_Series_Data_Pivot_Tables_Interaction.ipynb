<h3 align='Left'> Importing Libraries</h3>

```python
import pandas as pd
import pandas_datareader.data as web
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

<h3 align='Left'> Getting Data and Viewing with Pandas </h3>

```python
# https://pandas-datareader.readthedocs.io/en/latest/remote_data.html
google = web.DataReader('GOOG', data_source = 'google', start = '3/14/2009', end = '4/14/2016')
google.head()
```

<h3 align='Left'> Getting Data and Viewing with Pandas </h3>

```python
google_date = google.groupby(pd.TimeGrouper(freq='M')).mean()
google_date.head()
```

```python
google_date = google_date.reset_index()
google_date.head()
```

```python
pd.DataFrame(data = google_date.dropna())
```
